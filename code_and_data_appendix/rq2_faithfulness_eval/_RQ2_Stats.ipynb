```
import pandas as pd
import re
```

```
DATA_PATH = ""
```

```
df = pd.read_csv(DATA_PATH)

df = df.rename(columns={'XYZ ID':'student','Assessment ID':'assessment',\
                        'Conversation No.': 'turn','Role':'role','Response':'utterance'})
df = df[df['assessment'].isin([2,3,4])]

df.reset_index(drop=True,inplace=True)
```

```
df['group'] = (df['turn'] == 0).cumsum()
convos = [group_df.drop(columns='group') for _, group_df in df.groupby('group')]

print("Total Student-Agent Conversations:",len(convos))
print("Total Dialogue Turns:",len(df[df['role']=="assistant"]))
```

```
dfs_FA2 = [c for c in convos if c.iloc[0]['assessment']==2]
dfs_FA3 = [c for c in convos if c.iloc[0]['assessment']==3]
dfs_FA4 = [c for c in convos if c.iloc[0]['assessment']==4]
```

```
print("Total FA2 Student-Agent Conversations:",len(dfs_FA2))
print("Total FA3 Student-Agent Conversations:",len(dfs_FA3))
print("Total FA4 Student-Agent Conversations:",len(dfs_FA4))
```

```
print("Total FA2 Dialogue Turns:",pd.concat(dfs_FA2, ignore_index=True).query("role == 'assistant'").shape[0])
print("Total FA3 Dialogue Turns:",pd.concat(dfs_FA3, ignore_index=True).query("role == 'assistant'").shape[0])
print("Total FA4 Dialogue Turns:",pd.concat(dfs_FA4, ignore_index=True).query("role == 'assistant'").shape[0])
```
