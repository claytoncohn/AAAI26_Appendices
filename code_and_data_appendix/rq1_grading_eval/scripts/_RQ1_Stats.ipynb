```
import pandas as pd
```

```
DATA_PATH = ""
```

```
df_FA2 = pd.read_csv(DATA_PATH + "FA2.csv")
df_FA3 = pd.read_csv(DATA_PATH + "FA3.csv")
df_FA4 = pd.read_csv(DATA_PATH + "FA4.csv")
```

```
print("Total FAs:", len(df_FA2) + len(df_FA3) + len(df_FA4))
print("Total FA2:", len(df_FA2))
print("Total FA3:", len(df_FA3))
print("Total FA4:", len(df_FA4))
```

```
def score_distribution(df, max_points):
    s = df['score']
    return {
        "0 points": (s == 0).sum(),
        "partial credit": ((s > 0) & (s < max_points)).sum(),
        "max points": (s == max_points).sum()
    }
```

```
dist_FA2 = score_distribution(df_FA2, 9)
dist_FA3 = score_distribution(df_FA3, 5)
dist_FA4 = score_distribution(df_FA4, 4)

assert sum(dist_FA2.values()) == len(df_FA2)
assert sum(dist_FA3.values()) == len(df_FA3)
assert sum(dist_FA4.values()) == len(df_FA4)

summary = pd.DataFrame([dist_FA2, dist_FA3, dist_FA4],
                       index=["FA2", "FA3", "FA4"])
print(summary)
```

```
percent_summary = summary.div(summary.sum(axis=1), axis=0).mul(100).round(0)
print(percent_summary)
```
